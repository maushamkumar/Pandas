```python
students.head()
```

```python
# inner join => intersection of two data frames join on a common column
# Whenever you want to perform inner join first you should figure out a common column
# After that see if this will work for my code or not if yes then perform the inner join
students.merge(regs, on = 'student_id', how = 'inner') # First table merge and then second table
```

```python
courses.head()
```

```python
regs.head()
```

```python
# left join => all rows from the left table and only matching rows from the right table
courses.merge(regs, on = 'course_id', how = 'left')
```

```python
# right join => all rows from the right table and only matching rows from the left table
regs.merge(courses, on = 'course_id', how = 'right')
```

```python
# IPL problem
# find top 3 stadium with highest sixes per match
```

```python
# iformation about siz in ipl and stadium information in match
temp_df = ipl.merge(matches, left_on='match_id', right_on='id')
```

```python
six_df = temp_df[temp_df['batsman_runs'] == 6]
```

```python
# sixes per stadium
num_sixes = six_df.groupby('venue')['venue'].count()
```

```python
num_matches = matches['venue'].value_counts()
```

```python
(num_sixes/num_matches).sort_values(ascending=False)
```

```python
# find the orange cap holder of all the seasons
temp_df.groupby(['season', 'batsman'])['batsman_runs'].sum().reset_index().sort_values(by='batsman_runs', ascending=False).drop_duplicates(subset='season', keep='first')
```
