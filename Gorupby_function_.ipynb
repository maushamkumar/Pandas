```python
# Find number of movies starting with A for each group
genres.apply(lambda x: len(x[x['Series_Title'].str.startswith('A')]))
```

```python
# Find ranking of each movie for each group according to IMDB_Rating
def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

```python
# find normalised IMDB rating group wise

def normal(group):
    group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group

genres.apply(normal)
```

```python
# groupby on multiple cols
# size
# get_group

duo = movies.groupby(['Star1', 'Director'])
duo

# size
duo.size()

# get_group
duo.get_group(('Clint Eastwood', 'Clint Eastwood'))
```

```python
# Find the most earning actor -> director combo
duo['Gross'].sum().sort_values(ascending=False).head(1)
```

```python
# Find the best(in-terms of metasocre(avg)) actor -> genre combo
duo['Metascore'].mean().sort_values(ascending=False).head(1)

movies.groupby(['Star1', 'Genre'])['Metascore'].mean().reset_index().sort_values('Metascore', ascending=False)
```

```python
# agg on multiple groupby
duo = movies.groupby(['Star1', 'Director'])
# Specify the numeric columns to aggregate
numeric_cols = ['IMDB_Rating', 'Runtime']  # Replace with your actual numeric columns

# Perform the aggregation only on numeric columns
result = duo[numeric_cols].agg(['min', 'max', 'mean'])

result
```

```python
ipl = pd.read_csv('/content/deliveries.csv')
ipl.head()
```

```python
ipl.shape
```

```python
# find the top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# Find the batsman with max no of sixes
ipl[ipl['batsman_runs'] == 6].groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

```python
# find batsman with most number of 4s and 6s in last 5 over
# Create a new column to indicate boundary runs (4s and 6s)
ipl['run_by_boundaries'] = ipl['batsman_runs'].apply(lambda x: x if x in [4, 6] else 0)

# Filter for the last 5 overs (overs > 15)
last_5_overs = ipl[ipl['over'] > 15]

# Group by batsman and calculate the total number of 4s and 6s
boundary_counts = last_5_overs[last_5_overs['run_by_boundaries'] > 0].groupby('batsman')['run_by_boundaries'].count()

# Sort the result and get the top 10 batsmen
boundary_counts.sort_values(ascending=False).head(10)
```

```python
temp_df = ipl[ipl['over'] > 15]
temp_df = temp_df[(temp_df['batsman_runs'] == 6) | (temp_df['batsman_runs'] == 4)]
temp_df.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

```python
# find V kohli's record against all teams
v = ipl[ipl['batsman'] == 'V Kohli']
v.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```
